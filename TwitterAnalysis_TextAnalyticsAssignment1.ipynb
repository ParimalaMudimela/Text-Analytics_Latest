```python
#create word list for each token
word_list =[]
for record in data:
    #print(record)
    words = []
    tokens = record.split()
    #print(tokens)
    for token in tokens:
        words.append(token.lower())
    word_list.append(words)
word_list
```

```python
#Use Affinn model
#load Lexican file
lex_file = open("/Users/sree/Desktop/Parimala  MBA/AFINN-111.txt")
```

```python
lexicons = {}
records = lex_file.readlines()
```

```python

type(records)
```

```python

records[0]
```

```python
for record in records:
    #print(record)
    lexicons[record.rstrip('\n').split("\t")[0]] = int(record.rstrip('\n').split("\t")[1])
```

```python
#For each tweet
#For each word in tweet
#Get the word score (score is a number if the word is in Lexicon, 0 if not)
#Add all the scores and find the ploarity
#check strength of each tweet
strength = []
for tweet in word_list:
    #print(tweet)
    score = 0
    for word in tweet:
        if word in (lexicons):
            score = score + lexicons[word]
    strength.append(score)
```

```python
n = len(word_list)
print(n)
```

```python
#print strength for each tweet
for index in range(n):
    print(word_list[index], strength[index])
```

```python
#build a dataframe to compare the Lexican sentiment with original sentiment
orig_sentiment = df['Category']
lex_df_100 = pd.DataFrame({'Sentiment': strength, 'tweet': word_list, 'Orig_Sentiment': orig_sentiment})
lex_df_100.head(10)
```

```python
#Create a new column to store sentiment value based on Lexican model
lex_df_100['Sentiment_New'] = np.where(lex_df_100['Sentiment']<0,'Negative',(np.where(lex_df_100['Sentiment']==0,'Neutral','Postitive')))
lex_df_100.head(10)
```
